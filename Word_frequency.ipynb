```python
import nltk
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
file = open("scraping_and_segmentation/Audio_Data/CHILDES/Eng-NA/Gleason/Mother/andy/sentence_text.txt",newline='\n')
result = file.read()
words = word_tokenize(result)
```

```python
count_dict = dict()
word_set = set(words)
for w in word_set:
    freq_count = words.count(w)
    count_dict[w] = freq_count
```

```python
import operator
sorted_count_dict = list(reversed(sorted(count_dict.items(), key=operator.itemgetter(1))))
sorted_count_dict
```
